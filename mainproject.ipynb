```python
# the following line of code allows python to "import" or access outside material, in this case from the internet
import requests

# For the city input, only put the city name, with the first letter capitalized, with no state/country information. 
city = input("City: ")

# For the "dress code" input, make sure to enter in one of the three options with exact spelling and capitalization and no spaces
dc = input("Lazy, Casual, or Professional:")

#this line of code creates and stores the correct url for python to connect to and import data from
#the last part of the url 'format=j1' can be manipluated to change the format of the data
url = f"https://wttr.in/{city}?format=j1"

#this line tells python to request the data from our stored URL and convert into a python dictionary that is stored in the data variable
data = requests.get(url).json()

#the following lines get the data from our new python dictionary 
#it first goes to the current condition list from the database and by entering [0] is selects the first item in the list, which is another sublist
#then the code finds the correct item on the sublist to pull
#some items like description, rain, snow, and cloud are found in different places on the dictionary and require further steps to pull the correct data
#the numerical items are then changed to integers so they can be used in the if functions below
#the items are then stored in variables to be used later
temp = int(data["current_condition"][0]["temp_F"])
description = data["current_condition"][0]["weatherDesc"][0]["value"]
rain = int(data["weather"][0]["hourly"][0]["chanceofrain"])
wind = int(data["current_condition"][0]["windspeedMiles"])
uv = int(data["current_condition"][0]["uvIndex"])
snow = int(data["weather"][0]["hourly"][0]["chanceofsnow"])
cloud = int(data["weather"][0]["hourly"][0]["cloudcover"])

#all of the following lines of code are IF/THEN/ELSE functions that determine outfit selection based on the dress code and specific weather conditions

if temp < 32:
    if dc == "Lazy":
        extreme = "and layer up with a beanie, gloves & a coat."
    else:
        if dc == "Casual":
            extreme = "and layer up with a beanie, gloves & a parka."
        else:
            extreme = "and layer up awith gloves & a trench coat."
else:
    if temp > 85:
        extreme = "and dress in thin material if possible."
    else: extreme = "and enjoy the nice day."


if dc == "Lazy":
    if temp < 50:
        bottoms = "sweats,"
    else:
        bottoms = "shorts,"
else:
    if dc == "Casual":
        if temp < 75:
            bottoms = "jeans,"
        else:
            bottoms = "golf shorts,"
    else:
        bottoms = "slacks,"


if dc == "Lazy":
    if temp < 75:
        top = "a hoodie/sweatshirt,"
    else:
        top = "a t-shirt,"
else:
    if dc == "Casual":
        if temp < 50:
            top = "a sweater,"
        else:
            if temp < 75:
                top = "a jacket,"
            else:
                top = "a polo,"
    else:
        if temp < 50:
            top = "a blazer,"
        else:
            if temp < 75:
                top = "a button down,"
            else:
                top = "a short-sleeve button down,"


if rain < 30:
    raino = ""
else:
    if rain < 60:
        if dc == "Professional":
            raino = "bring an umbrella"
        else:
            raino = "bring a rain jacket"
    else:
        if dc == "Professional":
            raino = "use an umbrella"
        else:
            raino = "wear a rain jacket"


if dc == "Professional":
    windo = ""
else:
    if wind > 20:
        windo = "consider a wind breaker"
    else:
        windo = ""


if uv < 4:
    sun = ""
else:
    if uv < 7:
        sun = "consider wearing sunscreen"
    else:
        sun = "wear sunscreen"


if snow > 50:
    snowo = "wear snow boots"
else:
    snowo = ""

if cloud < 30:
    cloudo = "wear sunglasses and/or a hat."
else:
    if cloud < 60:
        cloudo = "bring sunwear."
    else:
        cloudo = "have fun today!"

# The following section is code than can print out the specific weather information if you need to check the accuracy or debug the code. 
# Simply remove the "#" to allow the code to run.

#print("Temperature (Â°F):", temp)
#print("Conditions:", description)
#print("Chance of Rain (%):", rain)
#print("Wind (MPH):", wind)
#print("UV:", uv)
#print("Chance of Snow (%):", snow)
#print("Cloud Coverage (%):", cloud)

# This is the final print function that creates the senetences to display the outfit recommendations
print("Right now in",city,"you should wear",bottoms,top,extreme,"You should also ",raino,windo,sun,snowo,cloudo)
```
