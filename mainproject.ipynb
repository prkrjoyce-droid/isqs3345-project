```python
import requests

city = input("City: ")
dc = input("Lazy, Casual, or Professional:")
url = f"https://wttr.in/{city}?format=j1"

data = requests.get(url).json()

temp_f = data["current_condition"][0]["temp_F"]
description = data["current_condition"][0]["weatherDesc"][0]["value"]
chance_of_rain = data["weather"][0]["hourly"][0]["chanceofrain"]
wind = data["current_condition"][0]["windspeedMiles"]
uv = data["current_condition"][0]["uvIndex"]
chance_snow = data["weather"][0]["hourly"][0]["chanceofsnow"]
cloud_cover = data["weather"][0]["hourly"][0]["cloudcover"]

print("Temperature (Â°F):", temp_f)
print("Conditions:", description)
print("Chance of Rain (%):", chance_of_rain)
print("Wind (MPH):", wind)
print("UV:", uv)
print("Chance of Snow (%):", chance_snow)
print("Cloud Coverage (%):", cloud_cover)
```
