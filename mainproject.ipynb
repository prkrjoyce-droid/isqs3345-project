```python
import requests

city = input("City: ")
dc = input("Lazy, Casual, or Professional:")
url = f"https://wttr.in/{city}?format=j1"

data = requests.get(url).json()

temp = data["current_condition"][0]["temp_F"]
description = data["current_condition"][0]["weatherDesc"][0]["value"]
rain = data["weather"][0]["hourly"][0]["chanceofrain"]
wind = data["current_condition"][0]["windspeedMiles"]
uv = data["current_condition"][0]["uvIndex"]
snow = data["weather"][0]["hourly"][0]["chanceofsnow"]
cloud = data["weather"][0]["hourly"][0]["cloudcover"]

print("Temperature (°F):", temp)
print("Conditions:", description)
print("Chance of Rain (%):", rain)
print("Wind (MPH):", wind)
print("UV:", uv)
print("Chance of Snow (%):", snow)
print("Cloud Coverage (%):", cloud)
```
