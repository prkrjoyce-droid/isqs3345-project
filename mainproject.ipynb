```python
import requests

city = input("City: ")
dc = input("Lazy, Casual, or Professional:")
url = f"https://wttr.in/{city}?format=j1"

data = requests.get(url).json()

temp = int(data["current_condition"][0]["temp_F"])
description = data["current_condition"][0]["weatherDesc"][0]["value"]
rain = int(data["weather"][0]["hourly"][0]["chanceofrain"])
wind = int(data["current_condition"][0]["windspeedMiles"])
uv = int(data["current_condition"][0]["uvIndex"])
snow = int(data["weather"][0]["hourly"][0]["chanceofsnow"])
cloud = int(data["weather"][0]["hourly"][0]["cloudcover"])

if temp < 32:
    if dc == "Lazy":
        extreme = "Layer up and bring a beanie, gloves  & a coat"
    else:
        if dc == "Casual":
            extreme = "Layer up and bring a beanie, gloves  & a parka"
        else:
            extreme = "Layer up and bring gloves  & a trench coat"
else:
    if temp > 85:
        extreme = "Dress in thin material if possible"
    else: extreme = ""


if dc == "Lazy":
    if temp < 50:
        bottoms = "sweats"
    else:
        bottoms = "shorts"
else:
    if dc == "Casual":
        if temp < 75:
            bottoms = "jeans"
        else:
            bottoms = "golf shorts"
    else:
        bottoms = "slacks"


if dc == "Lazy":
    if temp < 75:
        top = "hoodie/sweatshirt"
    else:
        top = "t-shirt"
else:
    if dc == "Casual":
        if temp < 50:
            top = "sweater"
        else:
            if temp < 75:
                top = "jacket"
            else:
                top = "polo"
    else:
        if temp < 50:
            top = "blazer"
        else:
            if temp < 75:
                top = "button down"
            else:
                top = "short-sleeve button down"


if rain < 30:
    raino = ""
else:
    if rain < 60:
        if dc == "Professional":
            raino = "bring an umbrella"
        else:
            raino = "bring a rain jacket"
    else:
        if dc == "Professional":
            raino = "use an umbrella"
        else:
            raino = "wear a rain jacket"


if dc == "Professional":
    windo = ""
else:
    if wind > 20:
        windo = "consider a wind breaker"
    else:
        windo = ""


if uv < 4:
    sun = ""
else:
    if uv < 7:
        sun = "consider wearing sunscreen"
    else:
        sun = "wear sunscreen"


if snow > 50:
    snowo = "wear snow boots"
else:
    snowo = ""

if cloud < 30:
    cloudo = "wear sunglasses/hat"
else:
    if cloud < 60:
        cloudo = "bring sunwear"
    else:
        cloudo = ""

print("Temperature (°F):", temp)
print("Conditions:", description)
print("Chance of Rain (%):", rain)
print("Wind (MPH):", wind)
print("UV:", uv)
print("Chance of Snow (%):", snow)
print("Cloud Coverage (%):", cloud)

print(extreme,bottoms,top,raino,windo,sun,snowo,cloudo)
```
