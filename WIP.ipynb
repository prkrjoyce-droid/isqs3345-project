```python
# Extremes WIP
temp = int(input("temp"))
dc = input ("dc")
if temp < 32:
    if dc == "Lazy":
        extreme = "Layer up and bring a beanie, gloves  & a coat"
    else:
        if dc == "Casual":
            extreme = "Layer up and bring a beanie, gloves  & a parka"
        else:
            extreme = "Layer up and bring gloves  & a trench coat"
else:
    if temp > 85:
        extreme = "Dress in thin material if possible"
    else: extreme = ""
print(extreme)
```

```python
# Bottoms WIP
temp = int(input("temp"))
dc = input ("dc")
if dc == "Lazy":
    if temp < 50:
        bottoms = "sweats"
    else:
        bottoms = "shorts"
else:
    if dc == "Casual":
        if temp < 75:
            bottoms = "jeans"
        else:
            bottoms = "golf shorts"
    else:
        bottoms = "slacks"
print(bottoms)
```

```python
# Tops WIP
temp = int(input("temp"))
dc = input ("dc")
if dc == "Lazy":
    if temp < 75:
        top = "hoodie/sweatshirt"
    else:
        top = "t-shirt"
else:
    if dc == "Casual":
        if temp < 50:
            top = "sweater"
        else:
            if temp < 75:
                top = "jacket"
            else:
                top = "polo"
    else:
        if temp < 50:
            top = "blazer"
        else:
            if temp < 75:
                top = "button down"
            else:
                top = "short-sleeve button down"
print(top)
```
